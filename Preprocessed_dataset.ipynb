```python
import pandas as pd
dataset = pd.read_csv("restaurant_data.csv")
```

```python
# Restaraunt name column is no needed,It don't give any improvement.
dataset = dataset.drop("Name",axis=1)
```

```python
dataset.dtypes
```

# Separate Qualitative data and Quantative data

```python
Quan = []
Qual = []
for i in dataset.columns:
    if (dataset[i].dtypes == "O"):
        Qual.append(i)
    else:
        Quan.append(i)
print(Quan)
print(Qual)
```

# Removing Outliers

```python
dataset.describe()
```

Create table for Outliers

```python
Outliers_Table = pd.DataFrame(index = ["IQR","Lesser_Outlier","Greater_Outliers","Min Value","Max Value"] , columns = Quan)
for i in Quan:
    Outliers_Table[i]["IQR"] = dataset.describe()[i]["75%"] - dataset.describe()[i]["25%"]
    Outliers_Table[i]["Lesser_Outlier"] = dataset.describe()[i]["25%"] - ( 1.5 * Outliers_Table[i]["IQR"] )
    Outliers_Table[i]["Greater_Outliers"] = dataset.describe()[i]["75%"] + ( 1.5 * Outliers_Table[i]["IQR"] )
    Outliers_Table[i]["Min Value"] = dataset.describe()[i]["min"]
    Outliers_Table[i]["Max Value"] = dataset.describe()[i]["max"]
```

```python
Outliers_Table 
```

Find out the outliers columns and save it in array

```python
Lesser = []
Greater = []
for i in Quan:
    if Outliers_Table[i]["Lesser_Outlier"] > Outliers_Table[i]["Min Value"] :
        Lesser.append(i)
    if Outliers_Table[i]["Greater_Outliers"] < Outliers_Table[i]["Max Value"] :
        Greater.append(i)
        
```

```python
print(Lesser)
print(Greater)
```

Replace original data to outlier data

```python
for j in Greater:
    for i in range(0,8368):
        if ( dataset[j][i] > Outliers_Table[j]["Greater_Outliers"]):
            dataset[j][i] = Outliers_Table[j]["Greater_Outliers"]
            
```

# Change catagorical data into numerical data 

```python
Preprocessed_data = pd.get_dummies(dataset,drop_first= True,dtype=int)
```
